```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Column names from the word document "Network Anamoly Detection" for the dataset
column_names = ["duration","protocol_type","service","flag","src_bytes","dst_bytes","land",
"wrong_fragment","urgent","hot","num_failed_logins","logged_in",
"num_compromised","root_shell","su_attempted","num_root","num_file_creations",
"num_shells","num_access_files","num_outbound_cmds","is_host_login",
"is_guest_login","count","srv_count","serror_rate", "srv_serror_rate",
"rerror_rate","srv_rerror_rate","same_srv_rate", "diff_srv_rate", "srv_diff_host_rate","dst_host_count","dst_host_srv_count","dst_host_same_srv_rate",
"dst_host_diff_srv_rate","dst_host_same_src_port_rate",
"dst_host_srv_diff_host_rate","dst_host_serror_rate","dst_host_srv_serror_rate",
"dst_host_rerror_rate","dst_host_srv_rerror_rate","attack", "last_flag"]

dataset = pd.read_csv("../Data/Train.txt", names=column_names)
```

```python
def scatter_plot(column, xlabel, ylabel, title):
    indexes = np.asarray(dataset['count'].keys())
    values = dataset[column].values.tolist()
    plt.scatter(x=indexes, y=values, s=1)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    plt.title(title)
    plt.show()

# Time related traffic features of each network connection vector
scatter_plot("count", "Index", "Number of connections",  "Number of similar connections to the same destination host in the past 2 seconds")
scatter_plot("srv_count", "Index", "Number of connections",  "Number of connections to the same service (port number) as the current connection in the past 2 seconds")

scatter_plot("serror_rate", "Index", "Percentage of connections",  "Percentage of connections with a 'flag' aggregated among connections in 'count'")
scatter_plot("srv_serror_rate", "Index", "Percentage of connections",  "Percentage of connections with a 'flag' aggregated among connections in 'srv_count'")

scatter_plot("rerror_rate", "Index", "Percentage of connections",  "Percentage of connections with the REJ 'flag' aggregated among connections in 'count'")
scatter_plot("srv_rerror_rate", "Index", "Percentage of connections",  "Percentage of connections with REJ 'flag' aggregated among connections in 'srv_count'")

scatter_plot("same_srv_rate", "Index", "Percentage of connections",  "Percentage of connections with the same service among the connections aggregated in 'count'")
scatter_plot("diff_srv_rate", "Index", "Percentage of connections",  "Percentage of connections that were different services among the connections aggregated in 'count'")
scatter_plot("srv_diff_host_rate", "Index", "Percentage of connections",  "Percentage of connections that were to different destination hosts among the aggregated connections in 'srv_count'")
```
