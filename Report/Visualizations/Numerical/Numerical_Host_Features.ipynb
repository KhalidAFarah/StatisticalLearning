```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Column names from the word document "Network Anamoly Detection" for the dataset
column_names = ["duration","protocol_type","service","flag","src_bytes","dst_bytes","land",
"wrong_fragment","urgent","hot","num_failed_logins","logged_in",
"num_compromised","root_shell","su_attempted","num_root","num_file_creations",
"num_shells","num_access_files","num_outbound_cmds","is_host_login",
"is_guest_login","count","srv_count","serror_rate", "srv_serror_rate",
"rerror_rate","srv_rerror_rate","same_srv_rate", "diff_srv_rate", "srv_diff_host_rate","dst_host_count","dst_host_srv_count","dst_host_same_srv_rate",
"dst_host_diff_srv_rate","dst_host_same_src_port_rate",
"dst_host_srv_diff_host_rate","dst_host_serror_rate","dst_host_srv_serror_rate",
"dst_host_rerror_rate","dst_host_srv_rerror_rate","attack", "last_flag"]

dataset = pd.read_csv("../../Data/Train.txt", names=column_names)
```

```python
def general_plot(column, xlabel, ylabel, title):
    indexes = np.asarray(dataset['count'].keys())
    values = dataset[column].values.tolist()
    plt.scatter(x=indexes, y=values, s=1)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    plt.title(title)
    plt.show()

# Host based traffic features in a network connection vector
# Number of connection with the same destination IP, and the number of connection the same source port
general_plot("dst_host_count", "Index", "Number of connections", "Number of connections with the same destination host IP address")
general_plot("dst_host_srv_count", "Index", "Number of connections", "Number of connection with the same port number")

# 
general_plot("dst_host_same_srv_rate", "Index", "Percentage of connections", "Percentage of connections that were to the same service among the aggregated connections in 'dst_host_count'")
general_plot("dst_host_diff_srv_rate", "Index", "Percentage of connections", "Percentage of connections that were to different services among the aggregated connections in 'dst_host_srv_count'")
general_plot("dst_host_same_src_port_rate", "Index", "Percentage of connections", "Percentage of connections that were to the same source port among the aggregated connections in 'dst_host_srv_count'")
general_plot("dst_host_srv_diff_host_rate", "Index", "Percentage of connections", "Percentage of connections that were to different destination hosts among the aggregated connections in 'dst_host_srv_count'")

# 
general_plot("dst_host_serror_rate", "Index", "Percentage of connections", "Percentage of connections that have an activated 'flag' among the aggregated connections in 'dst_host_count'")
general_plot("dst_host_srv_serror_rate", "Index", "Percentage of connections", "Percentage of connections that have an activated 'flag' among the aggregated connections in 'dst_host_srv_count'")
general_plot("dst_host_rerror_rate", "Index", "Percentage of connections", "Percentage of connections that have an activated REJ 'flag' among the aggregated connections in 'dst_host_count'")
general_plot("dst_host_srv_rerror_rate", "Index", "Percentage of connections", "Percentage of connections that have an activated REJ 'flag' among the aggregated connections in 'dst_host_srv_count'")
```
