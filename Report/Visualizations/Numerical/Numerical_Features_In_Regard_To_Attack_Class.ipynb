```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Column names from the word document "Network Anamoly Detection" for the dataset
column_names = ["duration","protocol_type","service","flag","src_bytes","dst_bytes","land",
"wrong_fragment","urgent","hot","num_failed_logins","logged_in",
"num_compromised","root_shell","su_attempted","num_root","num_file_creations",
"num_shells","num_access_files","num_outbound_cmds","is_host_login",
"is_guest_login","count","srv_count","serror_rate", "srv_serror_rate",
"rerror_rate","srv_rerror_rate","same_srv_rate", "diff_srv_rate", "srv_diff_host_rate","dst_host_count","dst_host_srv_count","dst_host_same_srv_rate",
"dst_host_diff_srv_rate","dst_host_same_src_port_rate",
"dst_host_srv_diff_host_rate","dst_host_serror_rate","dst_host_srv_serror_rate",
"dst_host_rerror_rate","dst_host_srv_rerror_rate","attack", "last_flag"]

dataset = pd.read_csv("../../Data/Train.txt", names=column_names)
dataset.head()

labels, values = np.unique(dataset['attack'], return_counts=True)
```

```python
dataset[dataset['attack'] == "normal"]['duration'].keys()
```

```python
def show_feature_in_relation_to_attack_bar_chart(labels, feature, xlabel, ylabel):
    for label in labels:
        title = f"Plot of the '{feature}' for the attack class '{label}'" 
        plt.scatter(dataset[dataset['attack'] == label][feature].keys(), dataset[dataset['attack'] == label][feature])
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)

        plt.title(title)
        plt.show()
```

```python
# Attack duration can very but so does normal connections.
show_feature_in_relation_to_attack_bar_chart(labels, "duration", "Duration index", "Duration")
```

```python
# Normal usually send fewer source bytes than the other attacks
show_feature_in_relation_to_attack_bar_chart(labels, "src_bytes", "Source bytes index", "Source byte value")
```

```python
# A lot of normal connection send a lot of destination bytes
show_feature_in_relation_to_attack_bar_chart(labels, "dst_bytes", "Destination bytes index", "Destination byte value")
```

```python
# Pod usually has one wrong fragment and teardrop usually has three wrong fragments.
show_feature_in_relation_to_attack_bar_chart(labels, "wrong_fragment", "Index", "Wrong fragments")
```

```python
# No significant findings
show_feature_in_relation_to_attack_bar_chart(labels, "urgent", "Index", "Urgent packets")
```
