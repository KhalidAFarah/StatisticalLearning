```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Column names from the word document "Network Anamoly Detection" for the dataset
column_names = ["duration","protocol_type","service","flag","src_bytes","dst_bytes","land",
"wrong_fragment","urgent","hot","num_failed_logins","logged_in",
"num_compromised","root_shell","su_attempted","num_root","num_file_creations",
"num_shells","num_access_files","num_outbound_cmds","is_host_login",
"is_guest_login","count","srv_count","serror_rate", "srv_serror_rate",
"rerror_rate","srv_rerror_rate","same_srv_rate", "diff_srv_rate", "srv_diff_host_rate","dst_host_count","dst_host_srv_count","dst_host_same_srv_rate",
"dst_host_diff_srv_rate","dst_host_same_src_port_rate",
"dst_host_srv_diff_host_rate","dst_host_serror_rate","dst_host_srv_serror_rate",
"dst_host_rerror_rate","dst_host_srv_rerror_rate","attack", "last_flag"]

dataset = pd.read_csv("../Data/Train.txt", names=column_names)
```

```python
def occurence_barchart_plot(column, xlabel, ylabel, title, graphs):
    labels, values = np.unique(dataset[column], return_counts=True)
    # print(len(labels))
    # print(len(values))
    # print(len(labels)/20)
    # print(round(len(labels)/20))
    labels_split = np.split(labels, graphs)
    values_split = np.split(values, graphs)
    
    for index in range(len(labels_split)):
        plt.barh(labels_split[index], values_split[index], align="center")
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.title(title)
        if len(labels_split) > 1:
            plt.title(f"{title} part {index+1}")
        else:
            plt.title(title)
        plt.show()

    print(f"{xlabel} and  their occurrences")
    for i in range(0, len(labels)):
        print(labels[i], "-",  values[i])
        
def occurence_scatter_plot(column, xlabel, ylabel, title):
    labels, values = np.unique(dataset[column], return_counts=True)
    plt.scatter(values, labels)
    # plt.yticks([1]*len(labels), labels=labels)
    # print([0.5]*2)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()

    print(f"{xlabel} and  their occurrences")
    for i in range(0, len(labels)):
        print(labels[i], "-",  values[i])

def occurence_pie_plot(column, xlabel, ylabel, title):
    labels, values = np.unique(dataset[column], return_counts=True)
    plt.pie(values, labels=labels)
    # plt.yticks([1]*len(labels), labels=labels)
    # print([0.5]*2)
    # plt.xlabel(xlabel)
    # plt.ylabel(ylabel)
    plt.title(title)
    # plt.legend(f"{column}: ")
    plt.show()

    print(f"{xlabel} and  their occurrences")
    for i in range(0, len(labels)):
        print(labels[i], "-",  values[i])
        
def occurence_stem_plot(column, xlabel, ylabel, title):
    labels, values = np.unique(dataset[column], return_counts=True)
    plt.pie(values, labels=labels)
    # plt.yticks([1]*len(labels), labels=labels)
    # print([0.5]*2)
    # plt.xlabel(xlabel)
    # plt.ylabel(ylabel)
    plt.title(title)
    plt.show()

    print(f"{xlabel} and  their occurrences")
    for i in range(0, len(labels)):
        print(labels[i], "-",  values[i])

#Nominal/categorical features
occurence_barchart_plot("protocol_type", "Occurrence", "Protocol types", "Bar chart of protocol types and their occurrences", 1)
occurence_barchart_plot("service", "Occurrence", "Service types", "Bar chart of service types and their occurrences", 2)
occurence_barchart_plot("flag", "Occurrence", "Flag types", "Bar chart of flag types and their occurrences", 1)
occurence_barchart_plot("attack", "Occurrence", "Attack class", "Bar chart of attack classes and their occurrences", 1)

#Binary features
#from the occurence of it is sufficiently little difference to the point of possibly disregarded, must be be investigated with the attack class
occurence_barchart_plot("land", "Occurrence", "Land", "Bar chart of instances where traffic is sent from the same machine and port and its occurrences", 1)
occurence_barchart_plot("logged_in", "Occurrence", "Logged in", "Bar chart of instances where the connection is logged in and its occurrences", 1)
occurence_barchart_plot("root_shell", "Occurrence", "Root shell", "Bar chart of instances where traffic is sent from the same machine and port and its occurrences", 1)
occurence_barchart_plot("su_attempted", "Occurrence", "Super user attempted", "Bar chart of instances where traffic is sent from the same machine and port and its occurrences", 1)
occurence_barchart_plot("is_host_login", "Occurrence", "Is logged in as host", "Bar chart of instances where traffic is sent from the same machine and port and its occurrences", 1)
occurence_barchart_plot("is_guest_login", "Occurrence", "Is logged in as guest", "Bar chart of instances where traffic is sent from the same machine and port and its occurrences", 1)
```

```python
def general_plot(column, xlabel, ylabel, title):
    plt.plot(dataset[column])
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    plt.title(title)
    plt.show()

general_plot("src_bytes", "Source bytes index", "Source byte value", "Histogram of source bytes")
general_plot("dst_bytes", "Destination bytes index", "Destination byte value",  "Histogram of destination bytes")

# Spiked point's value
print("Spiked largest source bytes points")
print(dataset["src_bytes"].nlargest(15))
print("Spiked largest destination bytes points")
print(dataset["dst_bytes"].nlargest(3))
```
