```python
import numpy as np
import scipy
import matplotlib.pylab as plt
```

Read the data

```python
ttphi, xxphi     = np.loadtxt('series-dphi.dat',   usecols=(0, 1), unpack=True)
tttheta, xxtheta = np.loadtxt('series-dtheta.dat', usecols=(0, 1), unpack=True)
```

```python
plt.plot(ttphi,xxphi)
plt.show()
plt.plot(tttheta,xxtheta)
plt.show()
```

### (1) Compute the autocorrelation of the data sets (separately):

```python
npttau         = 300
meanphi        = np.mean(xxphi)
varphi         = np.var(xxphi)
thisseries     = xxphi-meanphi
autocorrelacao = np.correlate(thisseries,thisseries,mode="full")
npt            = int(autocorrelacao.size/2)
plt.plot(autocorrelacao[npt:(npt+npttau)]/(varphi*npt),label = "autocorrelation series-dphi")
plt.show()

meantheta      = np.mean(xxtheta)
vartheta       = np.var(xxtheta)
thisseries     = xxtheta-meantheta
autocorrelacao = np.correlate(thisseries,thisseries,mode="full")
npt            = int(autocorrelacao.size/2)
plt.plot(autocorrelacao[npt:(npt+npttau)]/(vartheta*npt),label = "autocorrelation series-dtheta")
plt.show()
```

 as well as the correlation matrix of the two data sets together. Check <i>numpy.correlate</i>.

```python
correlacao = np.correlate(xxphi,xxtheta)
print("Correlation coefficient between both data series is:")
print(correlacao)
```

```python
matrixcorr = np.corrcoef(xxphi,xxtheta) 
print("Covariance matrix os both series:")
print(matrixcorr)
```

### (2) Taking file <i> series-dphi.dat</i> as the list of values of a variable $\phi$ and file <i>series-dtheta.dat</i> as the list of values of another variable $\theta$. Compute the joint histograma of the two variables.
