```python
import random
import matplotlib.pyplot as plt
import numpy as np
# Assignment 1
# number of samples
samples = 10**4

# Initial values
xval = []
# Generating 10^4 random values of x between 0 and 1
for i in range(0, samples):
    xval.append(random.uniform(0.0, 1.0))
    
plt.hist(xval, bins=20)
plt.show()
```

```python
histogram_values, bins_edges = np.histogram(xval)

new_histogram_values = np.zeros(len(histogram_values)+2)
bins_midpoints = np.zeros(len(histogram_values)+2)
for index in range(len(histogram_values)):
    new_histogram_values[index+1] =  histogram_values[index]
    bins_midpoints[index+1] =  0.5*(bins_edges[index] + bins_edges[index+1])

bins_midpoints[len(histogram_values) + 1] = bins_edges[len(histogram_values)]
new_histogram_values[len(histogram_values) + 1] = 0

plt.plot(bins_midpoints, new_histogram_values)
plt.show()
```

```python
# Assignment 2
cumulativ_histogram = np.zeros(len(histogram_values)+2)
norm = sum(histogram_values)

for i in range(len(histogram_values)):
    cumulativ_histogram[i+1] = cumulativ_histogram[i] + histogram_values[i]/np.float64(norm)

cumulativ_histogram[len(histogram_values)+1] = 1
plt.plot(bins_midpoints, cumulativ_histogram)
```
